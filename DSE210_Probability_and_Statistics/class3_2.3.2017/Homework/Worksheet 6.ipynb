Split the training set into two pieces - a training set of size 50000, and a separate validation set of size 10000. Also load in the test data.

```python
from sklearn.cross_validation import train_test_split
X_train, X_validation, y_train, y_validation = train_test_split(x, y, test_size=0.166666, random_state=1)
print 'The shape of the x train data set is: ', X_train.shape
print 'The shape of the y train data set is: ', y_train.shape
print 'The shape of the x validation data set is: ', X_validation.shape
print 'The shape of the y validation data set is: ', y_validation.shape
```

```python
X_test, y_test = loader.loadmnist('t10k-images.idx3-ubyte', 't10k-labels.idx1-ubyte')
print 'The shape of the x test data set is: ', X_test.shape
print 'The shape of the y test data set is: ', y_test.shape
```

## Part (c)
Now fit a Gaussian generative model to the training data of 50000 points:

    Determine the class probabilities: what fraction of pi_0 of the training points are digit 0, for instance? Call these values pi_0, ...., pi_9.
    
    Fit a Gaussian to each digit, by finding the mean and the covariance of the corresponding data points. 
    Let the Gaussian for the jth digit by Pj = N(mu_j, sigma_j).

Using these two pieces of information, you can classify new images x using Bayes' rule: simply pick the digit j for which pi_j,Pj(x) is largest.

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(X_train, y_train)
```

### Show the fraction (percentage) of images that belong to each class. Associated withe PI_j or the prior probabilities.

```python
priors = clf.class_prior_
length = len(priors)
for i in range(length):
    percentage = (priors[i])
    print 'The prior probability of image %s is' %i, '{0:.4f}'.format(percentage)
```

### Fit the train data set to the model used above (note: the validation set was not yet used/considered)

```python
classes = clf.classes_
posteriors = []

def class_grouping(class_id):
    grouping = []
    for i, group in enumerate(X_train):
        if y_train[i] == class_id:
            grouping.append(group)
    grouping = np.matrix(grouping)
    return grouping

for c in classes:
    grouping = class_grouping(c)
    mean = np.array(grouping.mean(0))[0]
    cov = np.cov(grouping.T)
    Px = multivariate_normal(mean, cov, allow_singular=True)
    posteriors.append(Px)
```

### Examining model predictions based on Bayes probability

```python
x = random.choice(X_test)

bayes_prob = []
for c in classes:
    prob = [c, priors[c] * posteriors[c].pdf(x)]
    bayes_prob.append(prob)

print 'The Bayes probability found is \n', bayes_prob
prediction = max(bayes_prob, key= lambda a: a[1])
print 'For this random example the model predicted a \n', y_train[prediction[0]]

print "Let's see how it compares to the actual image"
plt.figure(1, figsize=(3, 3))
plt.imshow(x.reshape(28, 28), cmap=plt.cm.gray_r, interpolation='nearest')
plt.show()
```

### Building the classifier

```python
Y = []
for x in X_test:
    bayes_prob = []
    for c in classes:
        prob = [c, priors[c] * posteriors[c].pdf(x)]
        bayes_prob.append(prob)
    prediction = max(bayes_prob, key= lambda a: a[1])
    Y.append(prediction[0])
```

```python
errors = (y_test != Y).sum()
total = X_test.shape[0]
print("Error rate: %d/%d = %f" %((errors, total, (errors/float(total)))))
```

### Our naively implemented Gaussian Classifer achieved a 9.8% success rate. Note such a low success rate can be due to using 'pdf' vs 'logpdf' which results in very small probabilities. Loss in probability precision results in inaccurate predictions.

### Comparing 100 of the predicted and actual classes for the digit data, the model predicted all of the digits to be 0s.

```python
def displayimage(image):
    plt.imshow(np.reshape(image, (28, 28)), cmap=plt.cm.gray)
    plt.axis('off')

indices = np.array(np.where((y_test != Y)==True))[0]
indices = indices[0:100]
index = 0
rows = len(indices)/10
cols = 10

plt.figure(figsize=(30,15))
for i in indices:
    index += 1
    plt.subplot(rows, cols, index)
    plt.subplots_adjust(hspace=.5)
    displayimage(X_test[i])
    plt.title('pre:%i, act:%i' %( Y[i], y_test[i]), fontsize = 20)
```

### Building the classifier: Using log probabilities

```python
Y = []
for x in X_test:
    bayes_prob = []
    for c in classes:
        prob = [c, np.log(priors[c]) + posteriors[c].logpdf(x)]
        bayes_prob.append(prob)
    prediction = max(bayes_prob, key= lambda a: a[1])
    Y.append(prediction[0])
```

```python
errors = (y_test != Y).sum()
total = X_test.shape[0]
print("Error rate: %d/%d = %f" %((errors, total, (errors/float(total)))))
```

### Our naively implemented Gaussian Classifer achieved a 81.4% success rate. Using logpdf resulted in more precise probabilities, thus a more accurate model.

### Comparing 100 of the predicted and actual classes for the digit data, the model predicted all of the digits to be 0s.

```python
indices = np.array(np.where((y_test != Y)==True))[0]
indices = indices[0:100]
index = 0
rows = len(indices)/10
cols = 10

plt.figure(figsize=(30,15))
for i in indices:
    index += 1
    plt.subplot(rows, cols, index)
    plt.subplots_adjust(hspace=.5)
    displayimage(X_test[i])
    plt.title('pre:%i, act:%i' %( Y[i], y_test[i]), fontsize = 20)
```

## Part (d)
One last step is needed: it is important to smooth the covariance matrices, and the usual way to do this is to add in cI, where c is some constant and I is the identity matrix. What value of c is right? Use the validation set to help you choose. That is, choose the value of c for which the resulting classifier makes the fewest mistakes on the validation set. What value of c did you get?

```python
smoothing_c = range(0, 5000, 500)
error_rates = []
for sc in smoothing_c:
    for c in classes:
        grouping = class_grouping(c)
        mean = np.array(grouping.mean(0))[0]
        cov = np.cov(grouping, rowvar=0)
        cov_smoothed = cov + (sc * np.eye(mean.shape[0]))
        p_x = multivariate_normal(mean, cov_smoothed, allow_singular=True)
        posteriors.append(p_x)

    Y = []
    for x in X_validation:
        bayes_prob = []
        for c in classes:
            prob = [c, np.log(priors[c]) + posteriors[c].logpdf(x)]
            bayes_prob.append(prob)
        prediction = max(bayes_prob, key= lambda a: a[1])
        Y.append(prediction[0])

    errors = (y_validation != Y).sum()
    total = X_validation.shape[0]
    error_rate = errors/float(total)
    error_rates.append(error_rate)
    print("Error rate for c= %s: %d/%d = %f" %((sc, errors, total, error_rate)))
```

```python
plt.plot(smoothing_c, error_rates)
plt.xlabel('c (smoothing constant)')
plt.ylabel('error rates')
plt.title('Error rate for varying smoothing constants')
```

### A c = 3500 contributed to creating the lowest error rate
