# PRACTICE QUESTION

1. Read the MNIST dataset. Split it into train and test data (80-20).
2. Perform PCA on the data and plot the test accuracy with increasing dimensions of the data. (Use the 1-nn classifier)

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('mnist_data.csv', names=range(784), header=None)
labels = pd.read_csv('mnist_labels.csv', names=['labels'])
data['labels'] = labels
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
acc_list = []

for i in range(1, 784):
    pca = PCA(n_components=i)
    X_train_new = pd.DataFrame(pca.fit_transform(X_train))
    X_test_new = pd.DataFrame(pca.transform(X_test))

    #Model
    clf = KNeighborsClassifier(1)
    clf.fit(X_train_new, y_train)
    pred = clf.predict(X_test_new)
    acc = accuracy_score(y_test, pred)
    print('For n_components = ', i, 'the accuracy is', acc)
    acc_list.append(acc)
```

```python
plt.plot(range(1, 784), acc_list)
plt.xlabel('Number of Components')
plt.ylabel('Test Accuracy')
plt.title('Test Accuracy vs. Number of Componenets for 784 Componenets')
```

```python
plt.plot(range(1, 100), acc_list[:99])
plt.xlabel('Number of Components')
plt.ylabel('Test Accuracy')
plt.title('Test Accuracy vs. Number of Componenets First 100 Componenets')
```

```python
plt.plot(range(1, 21), acc_list[:20])
plt.xlabel('Number of Components')
plt.ylabel('Test Accuracy')
plt.title('Test Accuracy vs. Number of Componenets First 20 Componenets')
```
